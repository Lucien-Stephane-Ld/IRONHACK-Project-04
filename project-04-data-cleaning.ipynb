## Data set information  


A team led by public policy professor Omar Asensio used a field experiment to collect data on 3,395 electric vehicle charging sessions between November 2014 and October 2015.

This dataset contains information from 3,395 high resolution electric vehicle charging sessions.  
The data contains sessions from 85 EV drivers with repeat usage at 105 stations across 25 sites at a workplace charging program.  
The workplace locations include facilities such as research and innovation centers, manufacturing, testing facilities and office headquarters for a firm participating in the U.S. Department of Energy (DOE) workplace charging challenge.  
The data is in a human and machine readable *.CSV format.  
The resolution of the data is to the nearest second, which is the same resolution as used in the analysis of the paper.

___

```python
df_clean = df.copy()
df_clean.head()
```

## Drop 'mon'-'sun' columns 

```python
df_clean.drop(columns=["Mon","Tues","Wed","Thurs","Fri","Sat","Sun"], inplace=True)
```

## Rename 'weekdays' to 'day' 

```python
df_clean.rename(columns={"weekday" : "Day"}, inplace=True)
```

## Round values of 'chargeTimeHrs' to two decimals

```python
df_clean['chargeTimeHrs'] = df_clean['chargeTimeHrs'].apply(lambda x: round(x, 2))

df_clean
```

## Rename 'chargeTimeHrs' to 'Charge Duration (h)'  

```python
df_clean.rename(columns={"chargeTimeHrs" : "Charge Duration (h)"}, inplace=True)

df_clean
```

## Drop 'created', 'ended' & 'reportedZip' columns

```python
df_clean.drop(columns=["created","ended","reportedZip"], inplace=True)
```

## Insert new column corresponding to Facility Code

```python
df_facilities
```

```python
def categorise(row):  
    if row['facilityType'] == 1:
        return 'Manufacturing'
    elif row['facilityType'] == 2:
        return 'Office'
    elif row['facilityType'] == 3:
        return 'R&D'
    return 'Others'
```

```python
df_clean['Facility Type'] = df.apply(lambda row: categorise(row), axis=1)

df_clean
```
